```python
# Import dependency
import pandas as pd
from pathlib import Path

# Import our input dataset
file_path = Path('C:/Users/kayem/Class/Crypto-Forecasting-/Resources/train.csv')
train_df = pd.read_csv(file_path)
train_df.head()
```

```python
# Import our input dataset
file_path_2 = Path('C:/Users/kayem/Class/Crypto-Forecasting-/Resources/asset_details.csv')
asset_details_df = pd.read_csv(file_path_2)
asset_details_df
```

```python
# Check the datatypes of your columns. 
train_df.dtypes
```

```python
# Convert the 'timestamp' column to datetime datatype.
train_df['timestamp'] = pd.to_datetime(train_df['timestamp'], unit='s')
train_df.head(10)
```

```python
# Merge asset_detail and train
clean_data = train_df.merge(asset_details_df, how='left', on="Asset_ID")
clean_data.head()
```

```python
# Reorder column to have Asset_ID after target
clean_data = clean_data.reindex(columns=['timestamp','Count','Open','High','Low','Volume','VWAP','Target','Asset_ID','Weight','Asset_Name'])
clean_data.head()
```

```python
# Identify incomplete rows
clean_data.count()
```

```python
# Drop all rows with missing information
clean_data = clean_data.dropna(how='any')
clean_data.head(10)
```

```python
# Verify dropped rows
clean_data.count()
```

```python
# Export the Dataframe as a new CSV file without the index.
clean_data.to_csv("clean_train.csv", index = False)
```
